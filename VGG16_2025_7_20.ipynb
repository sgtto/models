```python
from torch import nn
class VGG16(nn.module):
    def __init__(self):
        super(VGG16,self).__init__()
        #VGG有五层卷积层
        
        self.layer1=nn.Sequential(
            nn.Conv2d(in_channels=3,out_channels=64,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(64)#正则化
            nn.ReLU(inplace=True)
            nn.Conv2d(in_channels=64,out_channels=64,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(64)#正则化
            nn.ReLU(inplace=True)
            nn.MaxPool2d(kernel_size=2,stride=2)#out_size=16
            )
        
        self.layer2=nn.Sequential(
            nn.Conv2d(in_channels=64,out_channels=128,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(128)#正则化
            nn.ReLU(inplace=True)
            nn.Conv2d(in_channels=128,out_channels=128,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(128)#正则化
            nn.ReLU(inplace=True)
            nn.MaxPool2d(kernel_size=2,stride=2)#out_size=8
            )
        
        self.layer3=nn.Sequential(
            nn.Conv2d(in_channels=128,out_channels=256,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(256)#正则化
            nn.ReLU(inplace=True)
            nn.Conv2d(in_channels=256,out_channels=256,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(256)#正则化
            nn.ReLU(inplace=True)
            nn.MaxPool2d(kernel_size=2,stride=2)#out_size=4
            )
        
        self.layer4=nn.Sequential(
            nn.Conv2d(in_channels=256,out_channels=512,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(512)#正则化
            nn.ReLU(inplace=True)
            nn.Conv2d(in_channels=512,out_channels=512,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(512)#正则化
            nn.ReLU(inplace=True)
            nn.MaxPool2d(kernel_size=2,stride=2)#out_size=2
            )
            
        self.layer5=nn.Sequential(
            nn.Conv2d(in_channels=512,out_channels=512,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(512)#正则化
            nn.ReLU(inplace=True)
            nn.Conv2d(in_channels=512,out_channels=512,kernel_size=3,stride=1,padding1)
            nn.BatchNorm2d(512)#正则化
            nn.ReLU(inplace=True)
            nn.MaxPool2d(kernel_size=2,stride=2)#out_size=1
            )
        
        #将5层整合在一起
        self.conv=nn.Sequential(
            self.layer1
            self.layer2
            self.layer3
            self.layer4
            self.layer5
            )
        
        self.fc=nn.Sequential(
            nn.Linear(512,512)
            nn.ReLU(inplace=True)
            nn.Dropout(0.5)
            
            nn.Linear(512,256)
            nn.ReLU(inplace=True)
            nn.Dropout(0.5)
            
            nn.Linear(256,10)
            )
        
        def forward(self,x):
            self.conv(x)
            x=x.view(-1,512)
            self.fc(x)
            return x
        
```
